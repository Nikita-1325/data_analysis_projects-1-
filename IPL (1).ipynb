```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches = pd.read_csv(r"C:\Users\Nikita Saha\Documents\MATCHES.csv")
matches
```

```python
matches.drop(columns=["team1_playing11","team2_playing11","debut_players"],inplace = True)
```

```python
matches.head()
```

```python
matches.drop(columns=["mom_player"],inplace=True)
```

```python
batting = pd.read_csv(r"C:\Users\Nikita Saha\Documents\BATTING.csv")
```

```python
batting.head()
```

```python
bowling=pd.read_csv(r"C:\Users\Nikita Saha\Documents\BOWL.csv")
```

```python
bowling.head()
```

```python
#count of matches per season
matches_per_season=matches.groupby("season")["match_id"].count()
sns.barplot(x=matches_per_season.index,y=matches_per_season.values)
plt.title("Matches per season")
plt.xticks(rotation=45,ha='right')
plt.show
```

```python
#most successful teams
wins = matches["match_winner"].value_counts()
sns.barplot(x=wins.values, y=wins.index)
plt.title("Most successful teams")
plt.show
```

```python
#TOSS IMPACT
toss_winner = (matches["toss_winner"]== matches["match_winner"]).mean()#calculates the proportion
print(f"Toss winner also won the match:{toss_winner*100:.2f}%")
```

```python
#TOP RUN SCORERS
top_scorers = batting.groupby("batsman_id")["runs"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_scorers.values,y=top_scorers.index,palette="viridis")
plt.title('TOP SCORERS')
plt.xticks(rotation=45)
plt.show
```

```python
#TOP WICKET TAKERS
top_wickets = bowling.groupby("bowler_id")["wickets"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_wickets.values,y=top_wickets.index)
plt.title("TOP 10 WICKET TAKERS ")
plt.show()
```

```python
#top runs per season
runs_per_season = batting.groupby("match_id")["runs"].sum().head(10)
sns.barplot(x=runs_per_season.index,y=runs_per_season.values)
plt.title("Top Runs per season")
plt.xticks(rotation=45)
plt.show()
```

```python
#venue analysis
top_city = matches["match_venue_city"].value_counts().head(10)
sns.barplot(x=top_city.values,y=top_city.index)
plt.title("TOP CITIES HOSTING MOST MATCHES")
plt.xlabel("matches")
plt.show
```

```python
#team performnace
team_wins = matches["match_winner"].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=team_wins.values,y=team_wins.index)
plt.title("Most successful team")
plt.xlabel("wins")
plt.ylabel("Teams")
plt.show
```

```python
#Batsman performance

runs = batting.groupby("batsman_id")["runs"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
ax=sns.barplot(x= runs.index,y=runs.values)

plt.ylabel('Runs')
plt.xticks(rotation=45)
for i, v in enumerate (runs.values):
    ax.text(v+50,i,str(v),
    color="black",va="center",fontweight="bold")
plt.show()
```

```python
#total runs per matches

season_run=batting.groupby("match_id")['runs'].sum().reset_index()
season_run=season_run.merge(matches[["match_id","season"]], on = "match_id")
sns.barplot(x="season",y="runs",
data=season_run, estimator=sum,errorbar = None,color="orange")
plt.xticks(rotation=45)
plt.show()
```
